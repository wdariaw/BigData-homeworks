```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = [4, 16, 64, 128, 256, 400, 500, 600, 700,  800]
y = [0.008, 0.008384, 0.011488, 0.020672, 0.128352, 0.393664, 0.755744, 1.15389, 2.04778, 3.39789]
y_naive = [0.01856, 0.02352, 0.031008, 0.05376, 0.2664, 0.925344, 1.79027, 3.25846, 4.98819, 7.47459]

plt.figure(figsize=(12, 6))
plt.scatter(x, y, label='not naive')
plt.scatter(x, y_naive, label='naive')
plt.title(r"Зависимость времени работы методов $t$ от размера квадратных матриц $N$")
plt.xlabel(r"Размер матрицы $N$")
plt.ylabel(r"Время $t$")
plt.legend()
plt.show()
```
