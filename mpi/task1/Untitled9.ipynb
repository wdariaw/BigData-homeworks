```
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5, 6, 7, 8]
y_many = np.array([0.000401043, 0.00056626, 0.00342545, 0.00335201, 0.00349421, 0.00351469, 0.00351831, 0.00354979])
y_one = np.array([1.7777e-05, 1.7754e-05, 2.8576e-05, 3.9479e-05, 3.4863e-05, 2.7814e-05, 2.7743e-05, 3.8905e-05])

plt.plot(x, y_one / y_many)
plt.title("N = 1000")
plt.xlabel("Количество процессов")
plt.ylabel("Ускорение")
plt.show()
```
