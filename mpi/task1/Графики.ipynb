```
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5, 6, 7, 8]
y_many = np.array([0.00433632, 0.00422092, 0.00016566, 0.000269183, 0.000157748, 0.00313206, 0.00449439, 0.00654349])
y_one = np.array([3.159e-05, 3.2146e-05, 3.2502e-05, 3.2248e-05, 2.8821e-05,  2.7634e-05, 2.8035e-05, 2.9315e-05])

plt.plot(x, y_one / y_many)
plt.title("N = 1000")
plt.xlabel("Количество процессов")
plt.ylabel("Ускорение")
plt.show()
```
